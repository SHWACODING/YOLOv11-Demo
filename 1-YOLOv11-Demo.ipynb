```python
from ultralytics import YOLO
```

### Image Recognition ??

```python
## Load The YOLOv11 Recognition Model
model = YOLO("yolo11n.pt")
```

```python
model
```

```python
## Perform Object Detection On an Image
results = model("Data/children.jpeg")

results[0].show()
```

```python
## Perform Object Detection On a Video
results = model("Data/street.avi", save=True)
```

### Image Segmentation ??

```python
# Load YOLOv11 Segmentation Model
model = YOLO("yolo11n-seg.pt")
```

```python
# Perform Segmentation To an Image
results = model("Data/children.jpeg")

results[0].show()
```

### Pose Estimation ??

```python
# Load YOLOv11 Pose Estimation Model
model = YOLO("yolo11n-pose.pt")
```

```python
# Perform Pose Estimation On an Image
results = model("Data/children.jpeg")

results[0].show()
```

### Image Classification ??

```python
# Load YOLOv11 Image Classification Model
model = YOLO(model="yolo11n-cls.pt")

# Perform Image Classification On an Image
results = model("Data/cat.jpeg", save=True)

# Print the Results
print(results)
```
